```python
import pandas as pd
import seaborn as srn
import matplotlib.pyplot as plt
from google.colab import files
```

```python
# Carregamento da base de dados
upload = files.upload()
base = pd.read_csv('dados.csv', sep = ';')
base.head()
```

Sumario dos dados

```python
base['PIB'].describe()

```

Sumario de dados Valor Empenho

```python
base['VALOREMPENHO'].describe()
```

Boxplot e Histograma do PIB e valor empenho

```python
srn.boxplot(data=base.PIB, orient='v', showfliers=True).set_title('PIB')
```

```python
srn.histplot(base.PIB, color="blue", kde=False, stat="density", bins=10).set(title='Frequencia de PIB')
```

```python
srn.boxplot(data=base.VALOREMPENHO, orient='v', showfliers=True).set_title('Valor de Empenho')
```

```python
srn.histplot(base.VALOREMPENHO, color="blue", kde=False, stat="density", bins=10).set(title='Frequencia de Valor de empenho')
```

Maiores e Menores PIBS

```python
#Maior
MaiorPIB = base.sort_values('PIB', ascending=False).head(10)
MaiorPIB
MaiorPIB.plot.bar(x="MUNICIPIO", y="PIB",color='blue', title='10 Maiores PIBs')
```

```python
#Menor
MenorPIB = base.sort_values('PIB', ascending=True).head(10)
MenorPIB
MenorPIB.plot.bar(x="MUNICIPIO", y="PIB",color='red', title="10 Menores PIBs" )
```

Maiores e Menores Valor Empenho

```python
#Maior
MaiorVE = base.sort_values('VALOREMPENHO', ascending=False).head(10)
MaiorVE
MaiorVE.plot.bar(x="MUNICIPIO", y="VALOREMPENHO",color='blue', title='10 Maiores VALORES DE EMPENHO')
```

```python
#Menor
MenorVE = base.sort_values('VALOREMPENHO', ascending=True).head(10)
MenorVE
MenorVE.plot.bar(x="MUNICIPIO", y="VALOREMPENHO",color='red', title='10 Menores VALORES DE EMPENHO')
```

Calculando e vizualizando empenho/pib

```python
#Menor
MenorEmp = Base2.sort_values('EMPENHO', ascending=True).head(10)
MenorEmp
MenorEmp.plot.bar(x="MUNICIPIO", y="EMPENHO",color='red', title='10 Menores EMPENHO')
```
